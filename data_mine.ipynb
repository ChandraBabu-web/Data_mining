```python
os.environ['PATH'] += r"C://Users//chand//Downloads//chromedriver-win64//chromedriver.exe"
driver = webdriver.Chrome()
job_info = []

num_pages = 7


for page_num in range(1, num_pages + 1):
    if page_num == 1:
        driver.get("https://www.simplyhired.com/search?q=&l=Remote")
    else:
        
        next_page_url = f"https://www.simplyhired.com/search?q=&l=Remote&page={page_num}"
        driver.get(next_page_url)

    
    job_titles = driver.find_elements(By.TAG_NAME, "h2")
    company_elements = driver.find_elements(By.CSS_SELECTOR, "span[data-testid='companyName']")
    location_elements = driver.find_elements(By.CSS_SELECTOR, "span[data-testid='searchSerpJobLocation']")
    job_des = driver.find_elements(By.CSS_SELECTOR, "p[data-testid='searchSerpJobSnippet']")
    job_sal = driver.find_elements(By.CSS_SELECTOR, "p[data-testid='searchSerpJobSalaryConfirmed']")
    job_post = driver.find_elements(By.CSS_SELECTOR, "p[data-testid='searchSerpJobDateStamp']")
    job_link = driver.find_elements(By.XPATH, "//a[@class='chakra-button css-1djbb1k']")

    
    min_length = min(len(job_titles), len(company_elements), len(location_elements), len(job_des), len(job_sal), len(job_post), len(job_link))

    
    for i in range(min_length):
        job_title = job_titles[i].text
        company_name = company_elements[i].text
        job_location = location_elements[i].text
        job_desc = job_des[i].text
        job_sala = job_sal[i].text
        job_post_d = job_post[i].text
        job_links = job_link[i].get_attribute("href")

        
        job_info.append({
            "Job Title": job_title,
            "Company Name": company_name,
            "Job Location": job_location,
            "Job Description": job_desc,
            "Job Salary": job_sala,
            "Job Post": job_post_d,
            "Job Link": job_links
        })
job_info=job_info[:100]

for job in job_info:
    print("Job Title:", job["Job Title"])
    print("Company Name:", job["Company Name"])
    print("Job Location:", job["Job Location"])
    print("Job Description:", job["Job Description"])
    print("Job Salary:", job["Job Salary"])
    print("Job Post:", job["Job Post"])
    print("Job Link:", job["Job Link"])
    print("-" * 50)


driver.quit()
```

```python

csv_file_path = "Job_Postings_new.csv"

fieldnames = ["Job Title", "Company Name", "Job Location", "Job Description", "Job Salary", "Job Post", "Job Link"]

with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    for job in job_info:
        writer.writerow(job)

print("Job information has been written to", csv_file_path)
```

```python

job_links=df['Job Link']

driver = webdriver.Chrome()

qualifications_list = []

for link in job_links:
    driver.get(link)
    
    driver.implicitly_wait(5)
    
    
    page_source = driver.page_source
    
   
    soup = BeautifulSoup(page_source, 'html.parser')
    
  
    qualifications_section = soup.find('div', {'data-skipa11y': 'true', 'data-testid': 'viewJobBodyJobFullDescriptionContent'})
    
   
    if qualifications_section:
        qualifications =  qualifications_section.get_text()
        qualifications_list.append(qualifications)
    else:
        qualifications_list.append("Qualifications not found")


driver.quit()


for index, qualifications in enumerate(qualifications_list, start=1):
    print(f"Job {index} Full JD:")
    if isinstance(qualifications, list):
        for q in qualifications:
            print("-", q)
    else:
        print("-", qualifications)
    print("-" * 50)
```
